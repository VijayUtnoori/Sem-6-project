# 1.Importing Library

```python
import sys
import pandas as pd
import matplotlib
import numpy as np
import sklearn 
print("== library versions")
print(f"python :{sys.version.split()[0]}")
print(f"pandas :{pd.__version__}")
print(f"matplot :{matplotlib.__version__}")
print(f"numpy :{np.__version__}")
print(f"sklearn :{sklearn.__version__}")
```

# 2.Train the ML model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=[[1],[2],[3],[4],[5]]
y=[2,4,6,8,10]
model.fit(X,y)
user_value=int(input("enter the number"))
prediction = model.predict([[user_value]])
print("the prediction of I/P 6 is",prediction)
```

# 3.ploting

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,color="orange",edgecolor="black",label="prediction")
plt.plot(X,model.predict(X),color = 'blue',label = 'RegressionLine')
plt.title("regression line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

# 1.Activety Price Prediction

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [[1],[2],[3],[4],[5]]
y = [20,40,60,80,100]
model.fit(X,y)
user_value = int(input("enter the vlaue"))
prediction = model.predict([[user_value]])
print("the prediction I/P :",prediction)
```
