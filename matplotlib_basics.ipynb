# 3.bar chart
Bivariare analysis , numerical vs categorical , Usecase-Aggregate analysis of groups

```python
childerns = [10,20,30,10,25,40,60,10]
color = ["pink","black","orange","green","voilate","yellow","red","white"]
plt.bar(color, childerns, color=["orange"],edgecolor="black",label = "color")
plt.xticks(rotation=100)# its used for to print colors name vertical in bar gragh
plt.title("favtraite colors of student")
plt.xlabel("color")
plt.ylabel("childerns")
plt.legend()
plt.show()
```

```python
# for haorizontal bars
plt.barh(color, childerns, color="yellow", edgecolor="black",label="color")
plt.title("childerns fav color")
plt.xlabel("childerns")
plt.ylabel("color")
plt.show()
```

```python
data=pd.read_csv("batsman_season_record.csv")
print(data)
```

```python
plt.bar(np.arange(data.shape[0])-0.2 ,data['2015'],width=0.2)
plt.bar(np.arange(data.shape[0]),data['2016'],width=0.2)
plt.bar(np.arange(data.shape[0])+0.2,data['2017'],width=0.2)
plt.xticks(np.arange(data.shape[0]),data["batsman"])
plt.title("bar chart analysis of batsman performance in IPL")
plt.xlabel("batsman names")
plt.ylabel("Scored in IPL")
plt.show()
```

```python
# stacked bar chart
plt.bar(data["batsman"],data["2015"],label="2015",color="red")
plt.bar(data["batsman"],data["2016"],bottom=data["2015"],label="2016",color="orange") # BOTTOM work to go bottom on bar and another bar will be at the top 
plt.bar(data["batsman"],data["2017"],bottom=(data["2016"]+data["2015"]),label="2015",color="green")
plt.xticks(rotation=80)
plt.legend()
plt.show()
```

# 4.Histogram
Univariate Analysis , Applyed For Numerical col , Usecase- Frequent Count

```python
data=pd.read_csv("telecom_data.csv")
print(data)
```

```python
plt.hist(data["Age"],bins=(0,10,20,30,40,50,60,70),color="red",label="age")
plt.title("Age of users")
plt.xlabel("age group")
plt.ylabel("actual age")
plt.legend()
plt.show()
```

# 4.pie chart
Univariate/Bivariate , Categorical VS Numerical , Use case-To Find The Contribution Of Scale

```python
df=pd.read_csv("gayle-175.csv")
print(df)
```

```python
plt.pie(df["batsman_runs"],labels=df["batsman"],colors=["red","cyan","green","orange","pink","blue"],autopct="%0.1f%%")
plt.show()
```

```python
#explode to remove and highlight
plt.pie(df["batsman_runs"],labels=df["batsman"],autopct="%0.1f%%",explode=[0,0.1,0,0,0,0],shadow=True)
plt.show()
```
